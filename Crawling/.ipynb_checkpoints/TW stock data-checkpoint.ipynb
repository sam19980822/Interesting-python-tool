```python
def find_up_down(df):
    
    status = []
    for i in range(df.shape[0]):
        if i==0:
            status.append(0)
        
        elif df.iloc[i,6] > 1.02*df.iloc[i-1,4]:#如果當天的收盤價 > 前一日最高價，表示當日受前一日影響，漲價
            status.append(1)
        
        elif df.iloc[i,4]*1.02 < df.iloc[i-1,6]:  #如果當天的收盤價 < 前一日最高價，表示當日受前一日影響，跌價
            status.append(-1)
        else:
            status.append(0)
    df['status'] = status

    return df
```
